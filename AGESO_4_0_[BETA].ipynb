```python
############################## PATOLOGÍAS ##################################
from cie.cie10 import CIECodes #Importar librería CIE10
from IPython.display import clear_output

## ABRE E IMPORTA EL ARCHIVO "LibroP.xls" ##
filePath="/content/LibroP.xls"
openFile=xlrd.open_workbook(filePath)
hoja=openFile.sheet_by_index(0)

#Importar la información en una lista
patologias=[] #Aquí se guardan la información bruta del archivo "LibroP.xls"
for i in range(1, hoja.nrows):
  patologias.append(hoja.cell_value(i,0)) #hoja.cell_value(row,colum)

## Estandarizar los datos ##
patologias_unique = [] #Aquí se guardan las patologías individuales, con código e información inútil
for i in patologias:
  elementos = []
  if "|" in i:
    elementos = i.split("|") #Separa strings por "|"
    for j in elementos:
      patologias_unique.append(j)
  else:
    patologias_unique.append(i)

for k in patologias_unique:
  if k == "":
    patologias_unique.remove("") #Remueve elementos vacios

## Funciones axuliares ##
def get_code(patologia):
  n = patologia.find(":") #Encuentra la posición en el str de las patologías donde se encuentra ":"
  pat = patologia[:n] #Conserva los elementos del str hasta la posición del primer ":"
  return pat #str con el código de la patología

def get_patology_name(code):
  CIE = CIECodes() #Llama la librería e importa el diccionario con los códigos y nombres
  if CIE.info(code = code) is None: #Algunos códigos no están en la librería, aquí se verifica si están o no
    name = "None" #Si el código de patología no se encuentra a la librería, se le asigna nombre "None" a la patología con ese código
    print("Código no encontrado: ", code) #Imprime el código que no está en la librería
  else: #Si el código está en la librería
    name = CIE.info(code = code)["description"] #Obtiene el nombre de la patología asignada a dicho código de patología
  return name

def limpiar_jupyter():
    clear_output(wait=True)

#Obtiene los códigos de las patologías usando la función get_codes
codigos_patologias = []
no_codes = []
for i in range(len(patologias_unique)):
  if len(get_code(patologias_unique[i])) <= 5:
    codigos_patologias.append(get_code(patologias_unique[i]))
  else:
    no_codes.append(get_code(patologias_unique[i]))

#Buscar el nombre de la patología en la libreria CIE10 usando la función get_patology_name (Puede demorar un poco ejecutando)
patologias_final = []
#print("Buscando nombre de la patología en base de datos CIE10...")
for j in range(len(codigos_patologias)): #Bucle en "codigos_patologias" para usar la función "get_patology_name"
  #PORCENTAJE
  #P = int((j / len(codigos_patologias)) * 100)
  #c = 100 - P
  #if j%(int(len(codigos_patologias)/5)) == 0:
 #   print(P, "%")
  # FIN PORCENTAJE
  name = get_patology_name(codigos_patologias[j]) #Validar si todos los códigos tienen nombre en la libreria CIE10
  if name == "None":
    patologias_final.append(codigos_patologias[j]) #Se agregan a "patologias_final" con el nombre del código de la patología
  else:
    patologias_final.append(name) #Se agregan a "patologias_final" con el nombre de la patología asociada al código
#print("Hecho \n")
#limpiar_jupyter()

#Crear lista con patologías sin repetir
only_pat_names = set(patologias_final)
only_pat_names = list(only_pat_names)
patologia = only_pat_names

#Contar cuántas patogías hay
cantidad_patologia = []
for j in only_pat_names:
  cantidad_patologia.append(patologias_final.count(j))

#Imprimir
print("Patologías \n")
for i in range(len(patologia)-1):
  print(patologia[i],": ", cantidad_patologia[i])

#Crea archivo "Patologías.xls"
wb=openpyxl.Workbook()
pet=wb.active
for x in range(0, len(patologia)):
  pet.cell(row = x + 1, column = 1, value = patologia[x])
  pet.cell(row = x + 1, column = 2, value = cantidad_patologia[x])
wb.save('Patologías.xlsx')
################ FIN ##################
```
